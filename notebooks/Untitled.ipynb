```python
vote_df = pd.read_csv('../data/2016-vote-fips-acp.csv')
```

```python
electoral_df = pd.read_csv('../data/electoral_votes.csv')
```

```python
vote_by_state = pd.pivot_table(vote_df, index='st_name', values=['trump','clinton'], aggfunc=np.sum)
```

```python
vote_by_state.reset_index(inplace=True)
```

```python
vote_electoral = vote_by_state.merge(electoral_df, left_on="st_name", right_on="State", how="left" )
```
