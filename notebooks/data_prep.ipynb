### Handling Alaska

Turns out that vote totals aren't counted by the census's "county equivalents," and in the vote table were collected into a single fips code `02000` for the state (not by county).

```
print(votes_df[votes_df['st']=='AK'])

    county   fips  st  pct_report  total_votes          lead state.name  \
134    NaN  000AK  AK           1       309407  Donald Trump     Alaska   
135    NaN  02000  AK           1       309407  Donald Trump     Alaska   

```


[RRH elections](https://rrhelections.com/index.php/2018/02/02/alaska-results-by-county-equivalent-1960-2016/) explains the issues around the vote count, and makes estimates for each county equivalent, which I'm adding in here. See specifically [this file](https://drive.google.com/open?id=1Cc5AQunjTj03mQx0NVBpY3rQmXbPSWUT) at the bottom of the RHH elections page. 

````python
alaska_df = pd.read_csv('../data/alaska_county_equivalents.csv', dtype={'fips': object})
````

````python
alaska_df.head()
````

### Shannon County SD

Shannon County (FIPS 46113) was [renamed](https://www.ddorn.net/data/FIPS_County_Code_Changes.pdf) to Oglala Lakota County (FIPS 46102) as of May 1st, 2015.

````python
'46102' in votes_df.fips.values, '46102' in acp_df.fips.values
````

OK, so we have vote totals for Oglala Lakota County, so the fix here it to change the entry & FIPS code for Shannon County in the `acp_df` table to Oglala Lakota County. 

````python
acp_df.loc[acp_df['fips'] == '46113']
````

````python
acp_df.loc[2059, ['county']] = "Oglala Lakota SD"
acp_df.loc[2059, ['fips']] = "46102"
````

````python
acp_df.loc[2059]
````
